```python
# Top 10 countries vs covid 19 case
bar=px.bar(df.head(10),x='Country/Region',y='TotalCases',color='Country/Region',height=600,hover_data=['Country/Region','Continent'])
bar.update_layout(title='Top 10 Countries by Total COVID-19 Cases')
bar.show()
```

```python
# Country vs death case
bar2=px.bar(df.head(15),x='Country/Region',y='TotalDeaths', color='Country/Region',height=600,hover_data=['Country/Region','Continent'])
bar2.update_layout(title='Top 15 Countries by Total COVID-19 Deaths Cases')
bar2.show()
```

```python
# This chart visualizes COVID-19 recovery data across countries.
rcov=px.bar(df.head(20),x='TotalRecovered',y='Country/Region', color='TotalRecovered',height=600,hover_data=['Country/Region','Continent'])
rcov.update_layout(title='Top 20 Countries by Total COVID-19 Recovery Cases')
rcov.show()
```

```python
# Sort and select top 10 or 20 countries by TotalCases
df_sorted = df.sort_values('TotalCases', ascending=False).head(10)

# Melt the DataFrame to long format for grouped bars
df_melted = df_sorted.melt(
    id_vars=['Country/Region', 'Continent'],
    value_vars=['TotalCases', 'TotalRecovered'],
    var_name='CaseType',
    value_name='Count'
)

# Create grouped bar chart
fig = px.bar(
    df_melted,
    x='Country/Region',
    y='Count',
    color='CaseType',
    barmode='group',
    height=600,
    hover_data=['Continent']
)

fig.update_layout(title='Total Cases vs Total Recovered (Top 10 Countries)')
fig.show()
```

```python
# px.scatter, x,y, size, color, hover_data, size_max, log_x,log_y
# Creates an interactive scatter plot showing the top 30 countries based on total COVID-19 cases
sc=px.scatter(df.head(30),x='Country/Region',y='TotalCases',color='TotalCases',size='TotalCases',size_max=50,hover_data=['TotalDeaths','Continent'])
sc.update_layout(title='Top 30 countries based on total COVID-19 cases')
sc.show()
```

```python
# Creates a horizontal bar chart showing the top 10 entries from the dataset based on the number of Total COVID-19 Tests performed.
b=px.bar(df.head(10), x='TotalTests', y='Continent',
       color='TotalTests', orientation="h",height=500,
       hover_data=['Country/Region', 'Continent'])
b.update_layout(title='Total COVID-19 Tests performed.')
```

```python
#Creates an interactive scatter plot showing how COVID-19 Total Cases are distributed across different continents and countries.
sc1=px.scatter(df, x='Continent',y='TotalCases', hover_data=['Country/Region', 'Continent'],
           color='TotalCases', size='TotalCases', size_max=50)
sc1.update_layout(title='Total Cases')
```

```python
# This plot visualizes the top 20 countries by population using bubble size and position, while applying a logarithmic scale on the Y-axis for better readability of large differences.
sc2=px.scatter(df.head(20), x='Country/Region',y='Population', hover_data=['Country/Region', 'Continent'],
           color='Population', size='Population', size_max=30, log_y=True)
sc2.update_layout(title='Top 20 countries by population ')
```

```python
#Visualizes COVID-19 testing distribution across continents and countries, focusing on the top 50 records from the dataset.
total=px.scatter(df.head(50), x='Continent',y='TotalTests', hover_data=['Country/Region', 'Continent'],
           color='TotalTests', size='TotalTests', size_max=80, log_y=True)
total.show()
```

```python
# This scatter plot visualizes COVID-19 total deaths by continent, highlighting the top 50 countries with the highest death counts.
dt=px.scatter(df.head(50), x='Continent',y='TotalDeaths', hover_data=['Country/Region', 'Continent'],
           color='TotalDeaths', size='TotalDeaths', size_max=80, log_y=True)
dt.update_layout(title='Total death rate')
dt.show()
```

```python
# This chart visualizes COVID-19 total case counts for the top 100 countries, showing comparisons through bubble size and color.
tc=px.scatter(df.head(100), x='Country/Region', y='TotalCases',
           hover_data=['Country/Region', 'Continent'],
           color='TotalCases', size='TotalCases', size_max=80)
tc.update_layout(title='total cases')
tc.show()
```
